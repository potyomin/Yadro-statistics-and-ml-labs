## 1.1 Вероятностное пространство, формула Байеса

### Задача 1

Два события $A$ и $B$ называют независимыми, если
$$
P(A \cap B) = P(A)\cdot P(B).
$$
Если это не так, события зависимы.

#### (a) Несовместные события

Несовместные события — это такие, которые не могут произойти вместе.
Значит,
$$
A \cap B = \varnothing,\quad P(A \cap B) = 0.
$$

Если бы они были независимы и при этом имели ненулевые вероятности, то
$$
P(A \cap B) = P(A)\cdot P(B) > 0,
$$
получилось бы противоречие.

**Вывод:**
- если $P(A)>0$ и $P(B)>0$, несовместные события зависимы;
- независимыми они могут быть только в случае, когда одно из них вообще невозможно: $P(A)=0$ или $P(B)=0$.

#### (b) События, образующие $\sigma$-алгебру $\Sigma$

$\sigma$-алгебра $\Sigma$ — это просто набор событий, с которыми мы работаем.
То, что события лежат в одной $\Sigma$, само по себе ничего не говорит об их независимости.

В одной и той же $\sigma$-алгебре могут быть и независимые, и зависимые события.

**Пример.**
Пусть $\Omega = \{1,2,3,4\}$, все исходы равновероятны.

- $A=\{1,2\}$ и $B=\{1,3\}$:
  $P(A)=P(B)=\tfrac12$, $P(A\cap B)=\{1\}$, значит $P(A\cap B)=\tfrac14 = \tfrac12\cdot\tfrac12$.
  Эти события независимы.

- $C=\{1,2\}$ и $D=\{1,2,3\}$:
  $P(C)=\tfrac12$, $P(D)=\tfrac34$, $C\cap D = C$, значит $P(C\cap D)=\tfrac12 \neq \tfrac38$.
  Эти события зависимы.

Все четыре события входят в одну и ту же $\sigma$-алгебру.

**Вывод:**
про события, образующие $\Sigma$, нельзя сказать "всегда независимы" или "всегда зависимы" - бывает и так, и так.

#### (c) События, имеющие одинаковую вероятность

Пусть $P(A)=P(B)$. Для независимости нужно, чтобы
$$
P(A \cap B) = P(A)\cdot P(B).
$$
Но про $P(A \cap B)$ нам ничего не сказано, поэтому возможны оба варианта.

**Пример независимых событий.**
Тот же пример: $\Omega=\{1,2,3,4\}$, все исходы равновероятны.
$A=\{1,2\}$, $B=\{1,3\}$:
$P(A)=P(B)=\tfrac12$, $P(A\cap B)=\tfrac14 = \tfrac12\cdot\tfrac12$.
События независимы.

**Пример зависимых событий.**
В том же пространстве возьмём $C=\{1,2\}$ и $D=\{1,2\}$:
$P(C)=P(D)=\tfrac12$, но $C\cap D = C$ и $P(C\cap D)=\tfrac12 \neq \tfrac14$.
События зависимы.

**Вывод:**
равенство вероятностей $P(A)=P(B)$ не решает вопрос о зависимости.
Такие события могут быть как независимыми, так и зависимыми.

### Задача 3

**Условие.**
Производится выстрел по вращающейся круговой мишени, в которой закрашены два непересекающихся сектора с углом $20^\circ$.
Какова вероятность попадания в закрашенную область?

**Решение.**

Мишень вращается, поэтому можно считать, что угол, под которым пуля попадает в мишень, равномерно распределён от $0^\circ$ до $360^\circ$.
То есть все направления одинаково вероятны.

Полный угол круга:
$$
360^\circ.
$$

Закрашены два сектора по $20^\circ$, значит суммарный закрашенный угол:
$$
2 \cdot 20^\circ = 40^\circ.
$$

Вероятность попасть в закрашенную область — это доля закрашенного угла от полного круга:
$$
P = \frac{40^\circ}{360^\circ} = \frac{1}{9}.
$$

**Ответ:** $P = \dfrac{1}{9}$.


### Задача 4

Пусть первый пароход прибыл в момент времени $x$, второй — в момент времени $y$ ($x,y\in[0,24]$). Точка $(x,y)$ равновероятна в квадрате $0\le x\le24,\ 0\le y\le24$, его площадь
$$
S_{\text{общ}} = 24\cdot 24 = 576.
$$

Второму придётся ждать, если он пришёл в течение часа после первого:
$$
x < y < x + 1.
$$
Первому — если он пришёл в течение двух часов после второго:
$$
y < x < y + 2.
$$

На диаграмме это два невзаимно пересекающихся треугольника.

Площадь области без ожидания:
$$
S_{\text{без}} =
\frac{23\cdot 23}{2} + \frac{22\cdot 22}{2}
= \frac{529+484}{2}
= \frac{1013}{2} = 506{,}5.
$$

Площадь области, где есть ожидание:
$$
S_{\text{ожид}} = S_{\text{общ}} - S_{\text{без}}
= 576 - 506{,}5 = 69{,}5.
$$

Искомая вероятность:
$$
P = \frac{S_{\text{ожид}}}{S_{\text{общ}}}
= \frac{69{,}5}{576}
= \frac{139}{1152} \approx 0{,}121.
$$


### Задача 5

**Условие.**
Самолёт состоит из трёх частей:
(1) кабина и двигатель
(2) топливные баки
(3) планер.
Самолёт считается поражённым, если достаточно одного попадания в первую часть, или двух попаданий во вторую, или трёх попаданий в третью. Каждый из $m$ снарядов при попадании в самолёт попадает в первую часть с вероятностью $p_1$, во вторую — с вероятностью $p_2$, в третью — с вероятностью $p_3$ ($p_1+p_2+p_3=1$). Снаряды попадают в части независимо.

Известно, что в самолёт попало ровно $m$ снарядов. Найти условную вероятность $P(A \mid m)$ события $A$ = «самолёт поражён» для $m=1,2,3,4$.

Обозначим:

- $k_1$ — число попаданий в первую часть,
- $k_2$ — во вторую,
- $k_3$ — в третью, $k_1+k_2+k_3=m$.

Самолёт **поражён**, если
$$
k_1 \ge 1 \quad \text{или} \quad k_2 \ge 2 \quad \text{или} \quad k_3 \ge 3.
$$

Самолёт **не поражён**, если одновременно выполняется
$$
k_1 = 0,\quad k_2 \le 1,\quad k_3 \le 2.
$$

Вероятность конкретной тройки $(k_1,k_2,k_3)$ при $m$ снарядах:
$$
P(k_1,k_2,k_3) = \frac{m!}{k_1!k_2!k_3!}\,p_1^{k_1}p_2^{k_2}p_3^{k_3}.
$$

Тогда
$$
P(A\mid m) = 1 - P(\text{«самолёт не поражён»} \mid m).
$$

#### Случай $m = 1$

Самолёт не поражён, если единственное попадание идёт во 2-ю или 3-ю часть:

- $(0,1,0)$: вероятность $p_2$;
- $(0,0,1)$: вероятность $p_3$.

Отсюда
$$
P(\text{не поражён}\mid 1) = p_2 + p_3,
$$
$$
P(A\mid 1) = 1 - (p_2 + p_3) = p_1.
$$

#### Случай $m = 2$

Самолёт не поражён в двух ситуациях:

- оба попадания в третью часть: $(0,0,2)$, вероятность $p_3^2$;
- одно попадание во вторую и одно в третью: $(0,1,1)$, вероятность $2p_2p_3$.

Итого:
$$
P(\text{не поражён}\mid 2) = p_3^2 + 2p_2p_3,
$$
$$
P(A\mid 2) = 1 - (p_3^2 + 2p_2p_3) = 1 - p_3(2p_2 + p_3).
$$

#### Случай $m = 3$

Условия «не поражён»: $k_1=0$, $k_2 \le 1$, $k_3 \le 2$, $k_2+k_3=3$.
Единственный вариант: $(0,1,2)$.

Вероятность:
$$
P(0,1,2) = \frac{3!}{0!\,1!\,2!} p_2 p_3^2 = 3p_2p_3^2.
$$

Тогда
$$
P(\text{не поражён}\mid 3) = 3p_2p_3^2,
$$
$$
P(A\mid 3) = 1 - 3p_2p_3^2.
$$

#### Случай $m = 4$

Чтобы самолёт не был поражён, нужно
$$
k_1 = 0,\quad k_2 \le 1,\quad k_3 \le 2.
$$
Тогда $k_2 + k_3 \le 1 + 2 = 3$, но нам нужно $k_2 + k_3 = 4$.
Такой тройки $(k_1,k_2,k_3)$ **не существует**, значит

$$
P(\text{не поражён}\mid 4) = 0,\quad
P(A\mid 4) = 1.
$$

Итак,
$$
P(A\mid 1) = p_1,\quad
P(A\mid 2) = 1 - p_3(2p_2 + p_3),\quad
P(A\mid 3) = 1 - 3p_2p_3^2,\quad
P(A\mid 4) = 1.
$$


## 1.2 Случайный вектор и числовые характеристики

### Задача 1

**Условие.**
Пусть
$$
f_{\xi}(x,y) = \frac{e^{-2|y|}}{\pi(1 + x^2)}.
$$
Является ли эта функция плотностью распределения случайного вектора?

**Напоминание.**
Функция $f(x,y)$ является плотностью, если выполняются два условия:

1. $f(x,y) \ge 0$ для всех $x,y$;
2. $$\iint_{\mathbb{R}^2} f(x,y)\,dx\,dy = 1.$$

### 1. Неотрицательность

$e^{-2|y|} \ge 0$ для любого $y$, а $\pi(1+x^2) > 0$ для любого $x$.
Значит, дробь всегда неотрицательна:
$$
f_{\xi}(x,y) \ge 0.
$$

### 2. Интеграл по всей плоскости

Вычислим
$$
\iint_{\mathbb{R}^2} f_{\xi}(x,y)\,dx\,dy
= \iint_{\mathbb{R}^2} \frac{e^{-2|y|}}{\pi(1+x^2)}\,dx\,dy.
$$

Функция раскладывается в произведение по $x$ и $y$, поэтому
$$
\iint_{\mathbb{R}^2} f_{\xi}(x,y)\,dx\,dy
= \left(\frac{1}{\pi}\int_{-\infty}^{+\infty} \frac{dx}{1+x^2}\right)
  \left(\int_{-\infty}^{+\infty} e^{-2|y|}\,dy\right).
$$

Знаем стандартные интегралы:

1.
   $$
   \int_{-\infty}^{+\infty} \frac{dx}{1+x^2} = \pi.
   $$
2.
   $$
   \int_{-\infty}^{+\infty} e^{-2|y|}\,dy
   = 2\int_{0}^{+\infty} e^{-2y}\,dy
   = 2 \cdot \frac{1}{2} = 1.
   $$

Подставляем:
$$
\iint_{\mathbb{R}^2} f_{\xi}(x,y)\,dx\,dy
= \frac{1}{\pi}\cdot \pi \cdot 1 = 1.
$$

**Вывод.**
Оба условия выполнены: $f_{\xi}(x,y)\ge 0$ и интеграл равен 1.
Значит, данная функция является плотностью распределения случайного вектора.


### Задача 2

**Условие.**
Совместное распределение случайных величин $\xi$ и $\eta$ задано таблицей:

\begin{array}{c|ccc}
\xi \backslash \eta & -1 & 0 & 1 \\
\hline
-1 & \frac{1}{8} & \frac{1}{12} & \frac{7}{24} \\
1  & \frac{1}{3} & \frac{1}{6}  & 0
\end{array}

1. (a) Найти маргинальные распределения $\xi$ и $\eta$.
2. (b) Вычислить математические ожидания, ковариационную и корреляционную матрицы вектора $(\xi,\eta)$.
3. (c) Исследовать $\xi$ и $\eta$ на независимость и некоррелированность.

### (a) Маргинальные распределения

Для $\xi$ суммируем по строкам:

- $P(\xi=-1) = \frac18 + \frac1{12} + \frac7{24}
  = \frac{3}{24} + \frac{2}{24} + \frac{7}{24}
  = \frac{12}{24} = \frac12.$

- $P(\xi=1) = \frac13 + \frac16 + 0
  = \frac{2}{6} + \frac{1}{6}
  = \frac{3}{6} = \frac12.$

Итого:
$$
P(\xi=-1)=\frac12,\quad P(\xi=1)=\frac12.
$$

Для $\eta$ суммируем по столбцам:

- $P(\eta=-1) = \frac18 + \frac13
  = \frac{3}{24} + \frac{8}{24}
  = \frac{11}{24}.$

- $P(\eta=0) = \frac1{12} + \frac16
  = \frac{1}{12} + \frac{2}{12}
  = \frac{3}{12} = \frac14.$

- $P(\eta=1) = \frac7{24} + 0 = \frac7{24}.$

Итого:
$$
P(\eta=-1)=\frac{11}{24},\quad
P(\eta=0)=\frac14,\quad
P(\eta=1)=\frac{7}{24}.
$$

### (b) $E\xi$, $E\eta$, ковариация и корреляция

**Математические ожидания.**

Для $\xi$:
$$
E\xi = (-1)\cdot P(\xi=-1) + 1\cdot P(\xi=1)
     = (-1)\cdot\frac12 + 1\cdot\frac12 = 0.
$$

Для $\eta$:
$$
E\eta = (-1)\cdot\frac{11}{24} + 0\cdot\frac14 + 1\cdot\frac{7}{24}
      = \frac{-11+0+7}{24}
      = -\frac4{24} = -\frac16.
$$

**Вторые моменты и дисперсии.**

Для $\xi$:
$$
E\xi^2 = (-1)^2 \cdot \frac12 + 1^2 \cdot \frac12
       = \frac12 + \frac12 = 1,
$$
$$
\operatorname{Var}(\xi) = E\xi^2 - (E\xi)^2 = 1 - 0^2 = 1.
$$

Для $\eta$:
$$
E\eta^2 = (-1)^2 \cdot \frac{11}{24} + 0^2\cdot\frac14 + 1^2\cdot\frac{7}{24}
        = \frac{11}{24} + 0 + \frac{7}{24}
        = \frac{18}{24} = \frac34,
$$
$$
\operatorname{Var}(\eta)
= E\eta^2 - (E\eta)^2
= \frac34 - \left(-\frac16\right)^2
= \frac34 - \frac1{36}
= \frac{27-1}{36}
= \frac{26}{36}
= \frac{13}{18}.
$$

**Ковариация.**

Сначала найдём $E(\xi\eta)$, суммируя по всем клеткам таблицы:

\begin{aligned}
E(\xi\eta) &= (-1)(-1)\cdot\frac18
           + (-1)\cdot 0 \cdot\frac1{12}
           + (-1)\cdot 1 \cdot\frac7{24} \\
           &\quad + 1\cdot(-1)\cdot\frac13
           + 1\cdot 0 \cdot\frac16
           + 1\cdot 1 \cdot 0 \\
           &= \frac18 + 0 - \frac7{24} - \frac13 + 0 + 0 \\
           &= \frac{3}{24} - \frac{7}{24} - \frac{8}{24}
            = -\frac{12}{24} = -\frac12.
\end{aligned}


Тогда
$$
\operatorname{Cov}(\xi,\eta)
= E(\xi\eta) - E\xi\cdot E\eta
= -\frac12 - 0\cdot\left(-\frac16\right)
= -\frac12.
$$

Ковариационная матрица:
$\displaystyle
\Sigma =
\begin{pmatrix}
1 & -\tfrac12 \\
-\tfrac12 & \tfrac{13}{18}
\end{pmatrix}.
$

**Коэффициент корреляции.**

$$
\rho_{\xi,\eta}
= \frac{\operatorname{Cov}(\xi,\eta)}
       {\sqrt{\operatorname{Var}(\xi)\operatorname{Var}(\eta)}}
= \frac{-\frac12}{\sqrt{1\cdot \frac{13}{18}}}
= -\frac12 \sqrt{\frac{18}{13}}
\approx -0{,}59.
$$

Корреляционная матрица:
$\displaystyle
R =
\begin{pmatrix}
1 & \rho_{\xi,\eta} \\
\rho_{\xi,\eta} & 1
\end{pmatrix}
\approx
\begin{pmatrix}
1 & -0{,}59 \\
-0{,}59 & 1
\end{pmatrix}.
$

### (c) Независимость и некоррелированность

**Независимость.**
Для независимых величин должно выполняться
$$
P(\xi=x,\eta=y) = P(\xi=x)\,P(\eta=y)
$$
для всех пар $(x,y)$.

Проверим, например, точку $(\xi=1,\eta=1)$:

- по таблице: $P(\xi=1,\eta=1) = 0$;
- по произведению маргинальных:
  $P(\xi=1)\,P(\eta=1) = \frac12 \cdot \frac7{24} = \frac7{48} \ne 0$.

Равенство не выполняется, значит $\xi$ и $\eta$ **не независимы**.

**Некоррелированность.**
Некоррелированность означает $\operatorname{Cov}(\xi,\eta)=0$.
У нас
$$
\operatorname{Cov}(\xi,\eta) = -\frac12 \ne 0.
$$
Следовательно, $\xi$ и $\eta$ **коррелированы** (не являются некоррелированными).

**Итог:**

- маргинальные распределения:
  $$P(\xi=-1)=\tfrac12,\ P(\xi=1)=\tfrac12;$$
  $$P(\eta=-1)=\tfrac{11}{24},\ P(\eta=0)=\tfrac14,\ P(\eta=1)=\tfrac7{24};$$
- $E\xi=0,\ \operatorname{Var}(\xi)=1$;
- $E\eta=-\tfrac16,\ \operatorname{Var}(\eta)=\tfrac{13}{18}$;
- $\operatorname{Cov}(\xi,\eta)=-\tfrac12,\ \rho_{\xi,\eta}\approx -0{,}59$;
- величины $\xi$ и $\eta$ **не независимы** и **коррелированы**.


### Задача 3

**Условие.**
Есть два одинаковых тетраэдра с числами $1,2,3,4$ на гранях. Подкидываем оба и смотрим на выпавшие числа $\xi_1$ и $\xi_2$.

Задаём случайные величины:

- $\varphi_1 = \xi_1 + \xi_2$;
- $\varphi_2 = 1$, если одно из чисел делит другое ($\xi_1 \mid \xi_2$ или $\xi_2 \mid \xi_1$), и $\varphi_2 = 0$ иначе.

Всего $16$ равновероятных исходов $(\xi_1,\xi_2)$, каждый с вероятностью $1/16$.

#### (a) Совместное распределение $(\varphi_1,\varphi_2)$

Совместная таблица вероятностей (строки — значения $\varphi_1$, столбцы — значения $\varphi_2$):

$$
\begin{array}{c|cc}
\varphi_1 \backslash \varphi_2 & 0 & 1\\
\hline
2 & 0      & \frac{1}{16}\\
3 & 0      & \frac{2}{16}\\
4 & 0      & \frac{3}{16}\\
5 & \frac{2}{16} & \frac{2}{16}\\
6 & 0      & \frac{3}{16}\\
7 & \frac{2}{16} & 0\\
8 & 0      & \frac{1}{16}
\end{array}
$$

#### (b) Маргинальные распределения $\varphi_1$ и $\varphi_2$

Для $\varphi_1$ (суммируем по строкам):

- $P(\varphi_1=2)=\dfrac{1}{16}$,
- $P(\varphi_1=3)=\dfrac{2}{16}=\dfrac18$,
- $P(\varphi_1=4)=\dfrac{3}{16}$,
- $P(\varphi_1=5)=\dfrac{4}{16}=\dfrac14$,
- $P(\varphi_1=6)=\dfrac{3}{16}$,
- $P(\varphi_1=7)=\dfrac{2}{16}=\dfrac18$,
- $P(\varphi_1=8)=\dfrac{1}{16}$.

Для $\varphi_2$ (суммируем по столбцам):

- $P(\varphi_2=0)=\dfrac{4}{16}=\dfrac14$,
- $P(\varphi_2=1)=\dfrac{12}{16}=\dfrac34$.

#### (c) Математическое ожидание, дисперсии, ковариация и корреляция

**Математические ожидания**

$$
E\varphi_1 = 2\cdot\frac1{16}
+ 3\cdot\frac18
+ 4\cdot\frac3{16}
+ 5\cdot\frac14
+ 6\cdot\frac3{16}
+ 7\cdot\frac18
+ 8\cdot\frac1{16}
= 5.
$$

$$
E\varphi_2 = 0\cdot\frac14 + 1\cdot\frac34 = \frac34.
$$

**Дисперсии**

$$
E\varphi_1^2 = 2^2\cdot\frac1{16}
+ 3^2\cdot\frac18
+ 4^2\cdot\frac3{16}
+ 5^2\cdot\frac14
+ 6^2\cdot\frac3{16}
+ 7^2\cdot\frac18
+ 8^2\cdot\frac1{16}
= \frac{55}{2},
$$

$$
\operatorname{Var}(\varphi_1) = E\varphi_1^2 - (E\varphi_1)^2
= \frac{55}{2} - 25 = \frac{5}{2}.
$$

$$
\operatorname{Var}(\varphi_2) = \frac34 \cdot \frac14 = \frac{3}{16}.
$$

**Ковариация**

$$
E(\varphi_1\varphi_2)
= 2\cdot\frac1{16}
+ 3\cdot\frac2{16}
+ 4\cdot\frac3{16}
+ 5\cdot\frac2{16}
+ 6\cdot\frac3{16}
+ 8\cdot\frac1{16}
= \frac{7}{2},
$$

$$
\operatorname{Cov}(\varphi_1,\varphi_2)
= E(\varphi_1\varphi_2) - E\varphi_1\,E\varphi_2
= \frac{7}{2} - 5\cdot\frac34
= -\frac14.
$$

**Коэффициент корреляции**

$$
\rho_{\varphi_1,\varphi_2}
= \frac{\operatorname{Cov}(\varphi_1,\varphi_2)}
       {\sqrt{\operatorname{Var}(\varphi_1)\operatorname{Var}(\varphi_2)}}
= \frac{-\frac14}{\sqrt{\frac52\cdot\frac{3}{16}}}
= -\frac{2}{\sqrt{30}} \approx -0{,}37.
$$

#### (d) Независимость и некоррелированность

Проверка независимости:

$$
P(\varphi_1=2,\varphi_2=1) = \frac1{16},
\qquad
P(\varphi_1=2)\,P(\varphi_2=1)
= \frac1{16}\cdot\frac34
= \frac{3}{64} \ne \frac1{16}.
$$

Значит, $\varphi_1$ и $\varphi_2$ не независимы.

Ковариация
$$
\operatorname{Cov}(\varphi_1,\varphi_2) = -\frac14 \ne 0
$$
тоже не равна нулю, поэтому случайные величины коррелированы (не являются некоррелированными).


### Задача 4
**Условие.**
Пусть случайная величина $\xi$ равномерно распределена на отрезке $[-\pi,\pi]$:
$$
\xi \sim U[-\pi,\pi],\quad f_\xi(x)=\frac{1}{2\pi},\ x\in[-\pi,\pi].
$$
Определим
$$
\eta_1 = \cos\xi,\quad \eta_2 = \sin\xi.
$$

Найти $E\eta_1$, $E\eta_2$, $\operatorname{Var}(\eta_1)$, $\operatorname{Var}(\eta_2)$,
$\operatorname{Cov}(\eta_1,\eta_2)$, коэффициент корреляции и исследовать $\eta_1$ и $\eta_2$ на независимость и некоррелированность.

#### Математические ожидания

$$
E\eta_1 = E(\cos\xi) = \frac{1}{2\pi}\int_{-\pi}^{\pi} \cos x\,dx = 0,
$$
$$
E\eta_2 = E(\sin\xi) = \frac{1}{2\pi}\int_{-\pi}^{\pi} \sin x\,dx = 0.
$$

Из-за симметрии синуса и косинуса на отрезке $[-\pi,\pi]$ оба средних равны нулю.

#### Дисперсии

$$
E\eta_1^2 = E(\cos^2\xi) = \frac{1}{2\pi}\int_{-\pi}^{\pi} \cos^2 x\,dx.
$$

Используем формулу
$$
\cos^2 x = \frac{1+\cos 2x}{2}.
$$

Тогда
$$
E\eta_1^2 = \frac{1}{2\pi}\int_{-\pi}^{\pi} \frac{1+\cos 2x}{2}\,dx
= \frac{1}{4\pi}\left(\int_{-\pi}^{\pi}1\,dx + \int_{-\pi}^{\pi}\cos 2x\,dx\right)
= \frac{1}{4\pi}(2\pi + 0) = \frac12.
$$

Поэтому
$$
\operatorname{Var}(\eta_1) = E\eta_1^2 - (E\eta_1)^2 = \frac12.
$$

Аналогично для $\eta_2$:
$$
E\eta_2^2 = E(\sin^2\xi) = \frac12,\qquad
\operatorname{Var}(\eta_2) = \frac12.
$$

#### Ковариация и корреляция

Сначала найдём $E(\eta_1\eta_2)$:
$$
E(\eta_1\eta_2) = E(\cos\xi\sin\xi)
= \frac{1}{2\pi}\int_{-\pi}^{\pi} \cos x \sin x\,dx.
$$

Так как $\cos x\sin x = \frac12 \sin 2x$, имеем
$$
E(\eta_1\eta_2)
= \frac{1}{2\pi}\int_{-\pi}^{\pi} \frac12 \sin 2x\,dx
= \frac{1}{4\pi}\int_{-\pi}^{\pi} \sin 2x\,dx = 0.
$$

Отсюда
$$
\operatorname{Cov}(\eta_1,\eta_2)
= E(\eta_1\eta_2) - E\eta_1\cdot E\eta_2
= 0 - 0\cdot 0 = 0.
$$

Коэффициент корреляции:
$$
\rho_{\eta_1,\eta_2}
= \frac{\operatorname{Cov}(\eta_1,\eta_2)}
       {\sqrt{\operatorname{Var}(\eta_1)\operatorname{Var}(\eta_2)}}
= \frac{0}{\sqrt{\frac12\cdot\frac12}} = 0.
$$

Значит, $\eta_1$ и $\eta_2$ некоррелированы.

#### Независимость

Между $\eta_1$ и $\eta_2$ существует жёсткая связь:
$$
\eta_1^2 + \eta_2^2
= \cos^2\xi + \sin^2\xi = 1
$$
с вероятностью $1$. То есть все значения $(\eta_1,\eta_2)$ лежат на окружности единичного радиуса.

Если бы $\eta_1$ и $\eta_2$ были независимы и непрерывны, то вероятность того, что они точно лежат на одной и той же кривой, была бы равна $0$. Здесь же эта вероятность равна $1$.

Следовательно, $\eta_1$ и $\eta_2$ не независимы, хотя и имеют нулевую ковариацию.

**Итог:**

- $E\eta_1 = 0,\ \operatorname{Var}(\eta_1)=\dfrac12$;
- $E\eta_2 = 0,\ \operatorname{Var}(\eta_2)=\dfrac12$;
- $\operatorname{Cov}(\eta_1,\eta_2)=0,\ \rho_{\eta_1,\eta_2}=0$;
- случайные величины $\eta_1$ и $\eta_2$ некоррелированы, но не независимы.


### Задача 5

**Условие.**
Найти плотность распределения суммы двух независимых случайных величин $\xi$ и $\eta$, если
$$
\xi \sim \operatorname{Exp}_2,\quad \eta \sim U_{0,1}.
$$

То есть
$$
f_\xi(x) =
\begin{cases}
2e^{-2x}, & x \ge 0,\\
0, & x<0,
\end{cases}
\qquad
f_\eta(y) =
\begin{cases}
1, & 0 \le y \le 1,\\
0, & \text{иначе}.
\end{cases}
$$

Нужно найти плотность $f_Z(z)$ для $Z = \xi + \eta$.

#### Свёртка

Для суммы независимых непрерывных величин:
$$
f_Z(z) = \int_{-\infty}^{+\infty} f_\xi(t)\, f_\eta(z-t)\,dt.
$$

Ненулевые области:

- $f_\xi(t) \neq 0$ при $t \ge 0$;
- $f_\eta(z-t) \neq 0$, когда $0 \le z-t \le 1$, то есть $z-1 \le t \le z$.

Значит пределы интегрирования по $t$ — пересечение условий
$$
t \ge 0,\quad z-1 \le t \le z.
$$

#### 1) Случай $z < 0$

Сумма двух неотрицательных величин не может быть отрицательной, поэтому
$$
f_Z(z) = 0,\quad z<0.
$$

#### 2) Случай $0 \le z \le 1$

Здесь $z-1 \le 0$, поэтому нижняя граница берётся из $t \ge 0$:
$$
0 \le t \le z.
$$

Тогда
$$
f_Z(z) = \int_{0}^{z} 2e^{-2t}\,dt
= [-e^{-2t}]_{0}^{z}
= 1 - e^{-2z}, \quad 0 \le z \le 1.
$$

#### 3) Случай $z \ge 1$

Теперь $z-1 \ge 0$, так что
$$
z-1 \le t \le z.
$$

Получаем
$$
f_Z(z) = \int_{z-1}^{z} 2e^{-2t}\,dt
= [-e^{-2t}]_{z-1}^{z}
= e^{-2(z-1)} - e^{-2z}
= (e^{2}-1)e^{-2z}, \quad z \ge 1.
$$

#### Ответ

Плотность суммы $Z = \xi + \eta$ имеет вид
$$
f_Z(z) =
\begin{cases}
0, & z < 0,\\[4pt]
1 - e^{-2z}, & 0 \le z \le 1,\\[4pt]
(e^{2}-1)e^{-2z}, & z \ge 1.
\end{cases}
$$

Эта функция неотрицательна и её интеграл по всей прямой равен 1, поэтому это корректная плотность распределения.
